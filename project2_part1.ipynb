```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1.5,6.0,1.5,3.8,8.7,3.6,1.2,8.1,2.5,5,1.6,1.6,2.4,3.9,5.4])
y = np.array([3.1,5.7,1.7,3.6,8.7,3.2,1.0,8.0,2.4,6,1.1,1.3,2.4,3.9,4.8])
```

```python
a0 = 0        # y-intercept
a1 = 0        # slope
lr = 0.0001   # learning rate
iters = 100
errors = list()

for itr in range(iters):
    error_cost = 0
    pd_a0 = 0
    pd_a1 = 0
    for i in range(len(x)):
        y_pred_i = a0 + a1 * x[i]
        error_cost = error_cost + (y[i] - y_pred_i)**2
        for j in range(len(x)):
            pd_a0 = pd_a0 + (2 * ((a0 + a1 * x[j]) - y[j]))
            pd_a1 = pd_a1 + (2 * x[j] * ((a0 + a1 * x[j]) - y[j]))
        a0 = a0 - lr * pd_a0
        a1 = a1 - lr * pd_a1
    
    if itr % 100 == 0:
        print(itr, a0, a1)
    errors.append(error_cost)
```

```python
y_pred = a0 + a1*x
plt.scatter(x, y, color='blue')
plt.plot(x, y_pred, color='pink')
plt.xlabel("x")
plt.show("y")
```

# Gradient Descent 

```python
def f(x):
    return x**2 - 5*x

def df(x):
    return 2*x - 5

def gradient_descent(f, df, intial_guess, lr=0.01, epsilon=1e-6, max_iteration=1000):
    x = np.array(initial_guess)
    for itr in range(max_iteration):
        grad = df(x)
        if np.linalg.norm(grad) < epsilon:            
            break
        print(itr, grad)
        x = x - lr * grad
    return x, f(x)
```

```python
initial_guess = [5]

optimal_param, min_value = gradient_descent(f, df, initial_guess)

print(optimal_param, min_value)
```
